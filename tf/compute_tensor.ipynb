# linear algebra

## norm

```python
mat = tf.ones([2, 2], tf.float32)
res = tf.norm(mat)

sess.run(res)
```

```python
mat = tf.ones([2, 2], tf.float32)
res = tf.norm(mat, ord=1)

sess.run(res)
```

```python
mat = tf.ones([2, 2], tf.float32)
res = tf.norm(mat, ord=2)

sess.run(res)
```

```python
mat = tf.ones([2, 2], tf.float32)
res = tf.norm(mat, ord=np.inf)

sess.run(res)
```

```python
mat = tf.ones([2, 2], tf.float32)
res = tf.norm(mat, ord='fro')

sess.run(res)
```

https://stackoverflow.com/questions/44549658/tf-norm-error-valueerror-ord-must-be-a-supported-vector-norm-got-fro

```python
mat = tf.ones([2, 2], tf.float32)
res = tf.norm(mat, ord='fro', axis=(0,1))

sess.run(res)
```

## rank

```python
t = tf.constant([[[1, 1, 1], [2, 2, 2]], [[3, 3, 3], [4, 4, 4]]])
res = tf.rank(t)  # 3

sess.run(res)
```
